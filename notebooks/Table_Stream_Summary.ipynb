## Sort table properly

```python
# test
names_list = list(t['name'])

# Function to extract the prefix and number from the string
def sorting_key(s):
    # Use regex to split the string into a prefix (letters) and a number part
    match = re.match(r'([A-Za-z\s-]+)(\d+)', s)
    if match:
        prefix = match.group(1)  # Alphabetic part
        number = int(match.group(2))  # Numeric part
        return (prefix, number)
    else:
        return (s, 0)  # If no number, return the string with 0 as the number

# Sort using the custom sorting function
names_sorted = sorted(names_list, key=sorting_key)
ind_sorted = sorted(range(len(names_list)), key=lambda i: sorting_key(names_list[i]))

print(names_sorted)
```

## Write latex table

### Test setup

```python
tout = t[ind_sorted]
colnames = ['name', 'alt_name', 'length', 'width', 'dhelio', 'dgal', 'mstar', 'rperi', 'rapo', 'ecc', 'T', 'Etot', 'Lz', 
            'Lperp']
tout.keep_columns(colnames)
tout = tout[colnames]

new_colnames = dict(name='Stream', alt_name='Other name', length=r'$l$', width='$w$', dhelio=r'{$d_{h}$}',
                    dgal=r'$R_{gal}$', mstar=r'$M_\star$', rperi=r'$R_{peri}$', rapo='$R_{apo}$', ecc='{$e$}',
                    T='{$T_{orb}$}', Etot=r'{$E_{tot}$}', Lz=r'$L_z$', Lperp=r'$L_\perp$')
for k in new_colnames:
    tout.rename_column(k,new_colnames[k])
```

```python
tout.pprint_all()
```

```python
# generate latex table
latex_buffer = io.StringIO()

formats = {'Stream':'{}', 'Other name':'{}', r'$l$':'{:.1f}', '$w$':'{:.1f}', r'{$d_{h}$}':'{:.1f}',
           r'$R_{gal}$':'{:.1f}',
           r'$M_\star$':'{:7.1e}', r'$R_{peri}$':'{:.1f}', '$R_{apo}$':'{:.1f}', '{$e$}':'{:.2f}',
           r'{$T_{orb}$}':'{:.0f}', r'{$E_{tot}$}':'{:.3f}', r'$L_z$':'{:.1f}', r'$L_\perp$':'{:.1f}'}

tout.write(latex_buffer, format='ascii.latex', col_align='llSSSSSSSSrSSS',
           latexdict=dict(tabletype='', preamble=r'\centering', header_start=r'\hline\hline', 
                          data_start=r'\hline', data_end=r'\hline'),
           formats=formats)

# Remove the \begin{table} and \end{table} lines
latex_output = latex_buffer.getvalue()
lines = latex_output.splitlines()
filtered_lines = [line for line in lines if not line.startswith(r'\begin{}') and not line.startswith(r'\end{}')]

# Join the filtered lines back into a string
latex_output = "\n".join(filtered_lines)

# check table looks good
print(latex_output)
```

```python
# write to file
with open('../tex/tables/stream-summary.tex', 'w') as f:
    f.write(latex_output)
```

### Split tables across pages

```python
istop = [0, 80, len(t)]

colnames = ['name', 'alt_name', 'length', 'width', 'dhelio', 'dgal', 'mstar', 'rperi', 'rapo', 'ecc', 'T', 'Etot', 'Lz', 
                'Lperp']
new_colnames = dict(name='Stream', alt_name='Other name', length=r'$l$', width='$w$', dhelio=r'{$d_{h}$}',
                    dgal=r'$R_{gal}$', mstar=r'$M_\star$', rperi=r'$R_{peri}$', rapo='$R_{apo}$', ecc='{$e$}',
                    T='{$T_{orb}$}', Etot=r'{$E_{tot}$}', Lz=r'$L_z$', Lperp=r'$L_\perp$')
formats = {'Stream':'{}', 'Other name':'{}', r'$l$':'{:.1f}', '$w$':'{:.1f}', r'{$d_{h}$}':'{:.1f}',
               r'$R_{gal}$':'{:.1f}',
               r'$M_\star$':'{:7.1e}', r'$R_{peri}$':'{:.1f}', '$R_{apo}$':'{:.1f}', '{$e$}':'{:.2f}',
               r'{$T_{orb}$}':'{:.0f}', r'{$E_{tot}$}':'{:.3f}', r'$L_z$':'{:.1f}', r'$L_\perp$':'{:.1f}'}

for i in range(2):
    # split table
    tout = t[ind_sorted][istop[i]:istop[i+1]]
    tout.keep_columns(colnames)
    tout = tout[colnames]

    for k in new_colnames:
        tout.rename_column(k,new_colnames[k])
    
    # generate latex table
    latex_buffer = io.StringIO()

    tout.write(latex_buffer, format='ascii.latex', col_align='llSSSSSSSSrSSS',
               latexdict=dict(tabletype='', preamble=r'\centering', header_start=r'\hline\hline', 
                              data_start=r'\hline', data_end=r'\hline'),
               formats=formats)

    # Remove the \begin{table} and \end{table} lines
    latex_output = latex_buffer.getvalue()
    lines = latex_output.splitlines()
    filtered_lines = [line for line in lines if not line.startswith(r'\begin{}') and not line.startswith(r'\end{}')]

    # Join the filtered lines back into a string
    latex_output = "\n".join(filtered_lines)
    
    # write to file
    with open(f'../tex/tables/stream-summary-{i}.tex', 'w') as f:
        f.write(latex_output)
```
